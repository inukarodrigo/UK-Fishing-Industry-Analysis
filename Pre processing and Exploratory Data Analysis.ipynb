## Data Cleaning

###### Step 1: Removing duplicate values

```python
vessel_list_jan_feb_mar_subset = vessel_list_jan_feb_mar_subset.drop_duplicates()
vessel_list_jan_feb_mar_subset
```

###### Taking actions on null values

```python
# Lets have a look at the null values again
vessel_list_jan_feb_mar_subset.isna().sum()
```

```python
# Check how the data looks like in the rows with null values (Registered Tonnage)
vessel_list_jan_feb_mar_subset[vessel_list_jan_feb_mar_subset["Registered tonnage"].isna()]
```

###### Taking actions of null values under "Registered tonnage"

```python
# Identify the data points near the null value
length = 14.95
length_close_to_target_length = vessel_list_jan_feb_mar_subset[(vessel_list_jan_feb_mar_subset["Overall length"] >= length - 0.02) & (vessel_list_jan_feb_mar_subset["Overall length"] <= length + 0.02)]

# Plot
length_close_to_target_length[["Overall length", "Registered tonnage"]].plot(kind = "scatter", x = "Overall length", y = "Registered tonnage")
```

###### There are lots of tonnage values for the length of 14.95, therefore, we can get the mean value of those points to fill up the null value

```python
# Filtering out the rows of overall length of 14.95
rows_with_length_of_14_95 = vessel_list_jan_feb_mar_subset[vessel_list_jan_feb_mar_subset["Overall length"] == 14.95]

# Calculate the mean tonnage from the points with length of 14.95
mean_tonnage = rows_with_length_of_14_95["Registered tonnage"].mean()

vessel_list_jan_feb_mar_subset["Registered tonnage"].fillna(mean_tonnage, inplace = True)
```

##### The above code only change the subset of the dataset (A subset of the dataset was taken to remove the unnecessary columns and it is important to create a new dataset from the subset after cleaning.)

###### Taking actions on null values under "year build"

```python
# Calculate mean of the "Year built" column by "Administrative port"
mean_year_by_port = vessel_list_jan_feb_mar_subset.groupby("Administrative port")["Year built"].transform("mean")
# Rounding off the number
mean_year_by_port = mean_year_by_port.round()

vessel_list_jan_feb_mar_subset["Year built"].fillna(mean_year_by_port, inplace = True)
```

###### Removing the column "Country of construction" (It is not required to address the research question)

```python
vessel_list_jan_feb_mar_subset = vessel_list_jan_feb_mar_subset.drop("Country of construction", axis = 1)
```

```python
vessel_list_jan_feb_mar_subset.isna().sum()
```

###### Converting the sub categories under "Categories" to main categories

```python
mask_A_1 = vessel_list_jan_feb_mar_subset["Category"] == "CATEGORY A (PELAGIC)"
vessel_list_jan_feb_mar_subset.loc[mask_A_1, "Category"] = "CATEGORY A"

mask_A_2 = vessel_list_jan_feb_mar_subset["Category"] == "CATEGORY A (ISLAND VESSELS)"
vessel_list_jan_feb_mar_subset.loc[mask_A_2, "Category"] = "CATEGORY A"

vessel_list_jan_feb_mar_subset["Category"].value_counts()
```

###### Taking actions on Unknown FPO Code under "Fish producer organization" variable

```python
vessel_list_jan_feb_mar_subset[vessel_list_jan_feb_mar_subset["Fish producer organisation*"].isin(["Unknown FPO code (Historic data only)"])]
```

```python
# Changing the "Unknown FPO Code" to NON-SECTOR as it has the majority of data points
mask = vessel_list_jan_feb_mar_subset["Fish producer organisation*"] == "Unknown FPO code (Historic data only)"
vessel_list_jan_feb_mar_subset.loc[mask, "Fish producer organisation*"] = "NON-SECTOR"
```

### Removing outliers

```python
big_vessels = vessel_list_jan_feb_mar_subset[(vessel_list_jan_feb_mar_subset["Overall length"]>40) | (vessel_list_jan_feb_mar_subset["Vessel capacity units"]>700)]
big_vessels
```

###### The rows above have the potential to be outliers because the values are higher than usual values, one way to handle this is by replacing the extream values with mean, median or mode.

* It is important to find new mean, modes etc. to replace because the mean, mode that was taken past was affected by outliers.

```python
# Filtering the subset by removing outliers
vessel_list_jan_feb_mar_subset = vessel_list_jan_feb_mar_subset[(vessel_list_jan_feb_mar_subset["Overall length"]<=40) & (vessel_list_jan_feb_mar_subset["Vessel capacity units"]<=700)]

# Getting the stat values by grouping the category (That way, you can get better statistics)
stats_length = vessel_list_jan_feb_mar_subset.groupby("Category")["Overall length"].describe()
stats_length
```

```python
stats_length = vessel_list_jan_feb_mar_subset.groupby("Category")["Overall length"].describe().reset_index()[["Category","25%","75%"]]
stats_length
```

###### To handle the outliers, I will shrink the extream values to 75% quartile and the extram low values to 25% quartile

```python
# Merging the values of 25% and 75%
big_vessels = big_vessels.merge(stats_length, on = "Category", how = "left")
big_vessels
```

```python
# Replacing NaN values in the category type of "Blue whiting"
nan_row_index = big_vessels[(big_vessels['25%'].isna()) & (big_vessels['75%'].isna())].index
big_vessels.loc[nan_row_index, "25%"] = 86.4*25/100
big_vessels.loc[nan_row_index, "75%"] = 86.4*75/100
```

* I can now compare the minimum and maximum values with 25 and 75 quartiles and if it is outside the range, I will bring it in.

```python
mask = big_vessels["Overall length"] < big_vessels['25%']
big_vessels.loc[mask, "Overall length"] = big_vessels['25%']

mask = big_vessels["Overall length"] > big_vessels['75%']
big_vessels.loc[mask, "Overall length"] = big_vessels['75%']
```

```python
big_vessels
```

###### Doing the same thing to "Registered tonnage" and "Vessel capacity units" variables

```python
# Dropping the 25% and 75% columns
big_vessels = big_vessels.drop("25%", axis = 1)
big_vessels = big_vessels.drop("75%", axis = 1)

# Getting the descriptive statistics of "Registered tonnage" from the actual subset of the dataset
vessel_list_jan_feb_mar_subset.groupby("Category")["Vessel capacity units"].describe()
```

```python
# Filtering 25% and 75% quantiles
stats_capacity = vessel_list_jan_feb_mar_subset.groupby("Category")["Vessel capacity units"].describe().reset_index()[["Category","25%","75%"]]
stats_capacity
```

```python
# Merging it to the dataset and comparing and changing the values
big_vessels = big_vessels.merge(stats_capacity, on = "Category", how = "left")
```

```python
# Replacing NaN values in the category type of "Blue whiting"
nan_row_index = big_vessels[(big_vessels['25%'].isna()) & (big_vessels['75%'].isna())].index
big_vessels.loc[nan_row_index, "25%"] = 4777.92*25/100
big_vessels.loc[nan_row_index, "75%"] = 4777.92*75/100
```

```python
mask = big_vessels["Vessel capacity units"] < big_vessels['25%']
big_vessels.loc[mask, "Vessel capacity units"] = big_vessels['25%']

mask = big_vessels["Vessel capacity units"] > big_vessels['75%']
big_vessels.loc[mask, "Vessel capacity units"] = big_vessels['75%']
```

```python
big_vessels
```

###### Removing the row that has the category of "Blue Whiting" (Reason: The values even after shrinking are very higher and could affect the overall analysis)

```python
big_vessels = big_vessels[big_vessels["Category"] != "BLUE WHITING"]
```

###### Considering registered tonnage

```python
# Dropping the 25% and 75% columns
big_vessels = big_vessels.drop("25%", axis = 1)
big_vessels = big_vessels.drop("75%", axis = 1)

# Getting the descriptive statistics of "Registered tonnage" from the actual subset of the dataset
vessel_list_jan_feb_mar_subset.groupby("Category")["Registered tonnage"].describe()

# Filtering 25% and 75% quantiles
stats_tonnage = vessel_list_jan_feb_mar_subset.groupby("Category")["Registered tonnage"].describe().reset_index()[["Category","25%","75%"]]
stats_tonnage

# Merging it to the dataset and comparing and changing the values
big_vessels = big_vessels.merge(stats_tonnage, on = "Category", how = "left")

mask = big_vessels["Registered tonnage"] < big_vessels['25%']
big_vessels.loc[mask, "Registered tonnage"] = big_vessels['25%']

mask = big_vessels["Registered tonnage"] > big_vessels['75%']
big_vessels.loc[mask, "Registered tonnage"] = big_vessels['75%']
```

```python
big_vessels
```

```python
# Dropping the 25% and 75% columns
big_vessels = big_vessels.drop("25%", axis = 1)
big_vessels = big_vessels.drop("75%", axis = 1)
```

###### Now all the extream values were brought down to the 25% - 75% range. Now it is the time to check if there are outliers.

```python
# Combining big_vessels to the actual dataset
vessel_list_jan_feb_mar_subset = pd.concat([vessel_list_jan_feb_mar_subset, big_vessels], ignore_index = True)
```
